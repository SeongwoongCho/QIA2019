```python
import numpy as np
import os
from torchvision import transforms
import multiprocessing as mp
import librosa
import soundfile as sf
import wave

train_dir='../audioData/train/'
valid_dir='../audioData/val/'

is_coding=False

train_list_raw=os.listdir(train_dir)
valid_list_raw=os.listdir(valid_dir)

if is_coding:
    train_list_raw=train_list_raw[:150]
    valid_list_raw=valid_list_raw[:50]
frame_length = 0.025#25ms
frame_stride = 0.010
sr=44100

label2emo={'hap':0,'ang':1,'dis':2,'fea':3,'sad':4,'neu':5,'sur':6}
n_classes = 7


input_nfft = int(round(sr*frame_length))
input_stride = int(round(sr*frame_stride))

# train_data=[]
# valid_data=[]
# ###시간나면 Train,Valid 합치기
# def processforTrain(fileName):
#     if '.wav' in fileName:
#         try:
#             wavfile=wave.open(trainDir+fileName)
# #             print(fileName)
#             file=fileName.split('.')[0]
#             label=file.split('-')[-1]
#             y,sr=librosa.load(trainDir+fileName,sr=44100,mono=False)
#             y_mono=librosa.to_mono(y)
# #             print(fileName,sr ,type(y_mono),y_mono.shape)
#             S = librosa.feature.melspectrogram(y=y_mono,sr=sr,n_mels=128,n_fft=input_nfft, hop_length=input_stride)
#             train_data.append([S,label2emo[label]])
# #             print(type(label2emo[label]))
#         except Exception as ex:
#             print(fileName, ex)
# def processforValid(fileName):
#     if '.wav' in fileName:
#         file=fileName.split('.')[0]
#         label=file.split('-')[-1]
#         y,sr=librosa.load(valDir+fileName,sr=44100,mono=False)
#         y_mono=librosa.to_mono(y)
#         S = librosa.feature.melspectrogram(y=y_mono,sr=sr,n_mels=128, n_fft=input_nfft, hop_length=input_stride)
#         valid_data.append([S,label2emo[label]])
        
# pool=mp.Pool(14 if mp.cpu_count()>14 else mp.cpu_count())
# pool.map(processforTrain, trainList)
# pool.close()
# pool.join()
train_list=[]
for item in train_list_raw:
    if '.wav' in item:
        try:
            wave.open(train_dir+item)
        except Exception as ex:
            train_list.append(item)
valid_list=[]
for item in valid_list_raw:
    if '.wav' in item:
        try:
            wave.open(valid_dir+item)
        except Exception as ex:
            valid_list.append(item)

print("done")
print(len(train_list))
print(len(valid_list))
print(len(train_list_raw))
print(len(valid_list_raw))
        
        



```

```python
from torchvision import transforms
from PIL import Image
import random 
from torch.utils import data
class Dataset(data.Dataset):
    def __init__(self,file_list,root_dir,transform = None):
        self.file_list = file_list
        self.transform = transform
        self.root_dir=root_dir
    def __len__(self):
        return len(self.file_list)
    def __getitem__(self,idx):
            fileName = self.file_list[idx]
            wavfile=wave.open(self.root_dir+fileName)
            #print(self.root_dir+fileName)
            file=fileName.split('.')[0]
            label=file.split('-')[-1]
            y,sr=librosa.load(self.root_dir+fileName,sr=44100,mono=False)
            y_mono=librosa.to_mono(y)
            S = librosa.feature.melspectrogram(y=y_mono,sr=sr,n_mels=128,n_fft=input_nfft, hop_length=input_stride)
            #padding=np.zeros((128,1500))
            #padding[:S.shape[0],:S.shape[1]]=S
            tfms = transforms.Compose([
                transforms.Resize(224),
                transforms.RandomCrop(224),
                transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),])
            img = tfms(Image.fromarray(S,"RGB"))
            return img, label2emo[label]
```

```python
import torch.nn as nn
import csv
from torch import optim
from tqdm import tqdm_notebook, tqdm
import torch
from efficientnet_pytorch import EfficientNet
import torch.utils.data as data
from apex import amp

criterion=nn.CrossEntropyLoss().cuda()
learning_rate=3e-4
n_epoch=50

batch_size=24
num_workers= 14 if mp.cpu_count()>14 else mp.cpu_count()-1

save_path="./model/EffNET_1"
os.makedirs(save_path,exist_ok=True)

model=EfficientNet.from_name(model_name='efficientnet-b0',
                             override_params={'num_classes':n_classes})

model.cuda()


train_loader=data.DataLoader(dataset=Dataset(file_list=train_list,root_dir=train_dir),batch_size=batch_size,
                        num_workers=num_workers,shuffle=True)
print("train_loader",train_loader)
valid_loader=data.DataLoader(dataset=Dataset(file_list=valid_list,root_dir=valid_dir),batch_size=batch_size,
                        num_workers=num_workers,shuffle=True)


optimizer=optim.Adam(model.parameters(),lr=learning_rate)
model, optimizer = amp.initialize(model, optimizer, opt_level="O1")
model=nn.DataParallel(model)
print("training....")
log=open(os.path.join(save_path,'log.csv'),'w', encoding='utf-8',newline='')
log_writer=csv.writer(log)

best_val=np.inf
for epoch in range(n_epoch):
    train_loss=0
    train_acc=0
    
    optimizer.zero_grad()
    model.train()
    for idx,(x_train,y_train) in enumerate(tqdm_notebook(train_loader)):
        x_train,y_train=x_train.cuda(),y_train.type(torch.LongTensor).cuda()
        pred=model(x_train)
        loss=criterion(pred,y_train)
        with amp.scale_loss(loss, optimizer) as scaled_loss:
            scaled_loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        train_loss+=loss.item()/len(train_loader)
        pred=pred.detach().max(1)[1]
        train_acc+=pred.eq(y_train.view_as(pred)).sum().item()
    train_acc/=len(train_list)
    valid_loss=0
    valid_acc=0
    
    model.eval()
    with torch.no_grad():
        for idx,(x_val,y_val) in enumerate(tqdm_notebook(valid_loader)):
            x_val,y_val=x_val.cuda(),y_val.type(torch.LongTensor).cuda()
            pred=model(x_val)
            loss=criterion(pred,y_val)

            valid_loss+=loss.item()/len(valid_loader)
            
            pred=pred.detach().max(1)[1]
            valid_acc+=pred.eq(y_val.view_as(pred)).sum().item()
        valid_acc/=len(valid_list)
    if valid_loss<best_val:
        torch.save(model.state_dict(),os.path.join(save_path,'best.pth'))
        best_val=valid_loss
        
    log_writer.writerow([epoch,train_loss,train_acc,valid_loss,valid_acc])
    log.flush()
    print("Epoch [%d]/[%d] train_loss: %.6f train_Acc: %.6f valid_loss: %.6f valid_acc:%.6f"%(
    epoch,n_epoch,train_loss,train_acc,valid_loss,valid_acc))
    
log.close()
```
