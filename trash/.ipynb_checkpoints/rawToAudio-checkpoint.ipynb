```python
import multiprocessing as mp
from moviepy.editor import *
import os
import sys

fileList=os.listdir('../qia2019-1/test2')
dir='../qia2019-1/test2/'
targetDir='../audioData/test2/'
# fftDir='../fftData512_1024/val/'
# hop_length=512
# window_size=1024
os.makedirs(targetDir,exist_ok=True)
def process(file):
    if '.mp4' in file:
        filename=file.split('.')[0]
        fd=open(targetDir+filename+".wav","w")

        video=VideoFileClip(dir+file)
        audio=video.audio
        audio.write_audiofile(targetDir+filename+".wav")

        fd.close()
pool=mp.Pool(mp.cpu_count())
pool.map(process, fileList)
pool.close()
pool.join()
        
```

```python
import multiprocessing as mp
from moviepy.editor import *
import os
import sys

def process(arg):
    dir,file,targetDir,targetFile = arg
    if '.mp4' in file:
        filename=file.split('.')[0]
        fd=open(os.path.join(targetDir,targetFile)+".wav","w")

        video=VideoFileClip(os.path.join(dir,file))
        audio=video.audio
        audio.write_audiofile(os.path.join(targetDir,targetFile)+".wav")

        fd.close()

dirList=os.listdir('../../ourdata/')
print(dirList)
args = []
for dir in dirList:
    if not '.' in dir:
        for file in os.listdir('../../ourdata/' + dir):
            targetFile = file[:-4] + "-" + dir
            _dir = os.path.join('../../ourdata/',dir)
            targetDir='../../audioData/test_private/'
            args.append((_dir,file,targetDir,targetFile))
        
os.makedirs('../../audioData/test_private/',exist_ok=True)

pool=mp.Pool(mp.cpu_count())
pool.map(process, args)
pool.close()
pool.join()
```
